```python
import pandas as pd;
import sweetviz as sv;
```

```python
# 1. Cargo el dataset
try:
    # Intentamos primero con la configuración estándar europea
    df = pd.read_csv('../data/raw/dataset_accidentes_de_trabajo.csv', sep=';', encoding='utf-8')
except UnicodeDecodeError:
    # Si falla por los acentos/eñes, usamos latin1
    df = pd.read_csv('../data/raw/dataset_accidentes_de_trabajo.csv', sep=';', encoding='latin1')

# Vemos las primeras 5 filas y la estructura
display(df.head())
print(df.info())
```

```python
# 2. Generar el reporte mágico
reporte = sv.analyze(df, pairwise_analysis='off')
reporte.show_html('../reports/accidentes_laborales.html')
```

```python
df.rename(columns={'Unnamed: 0': 'Periodo'}, inplace=True)
```

```python
# 3. "Derretir" (Melt) la tabla para hacerla amigable para Google Sheets / Tableau
df_long = pd.melt(
    df,
    id_vars=['Periodo'], 
    var_name='Categoria_Sector_Sexo', 
    value_name='Total_Accidentes'
)
```

```python
df_long['Total_Accidentes'] = pd.to_numeric(df_long['Total_Accidentes'], errors='coerce')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Preparar los datos específicos que mencionaste
# Extraemos los nombres de las columnas (simplificamos los nombres largos para que quepan en el gráfico)
sectores = ['Construcción', 'Suministro de Energía']
promedios = [343.17, 2.28] # Los datos que sacaste de tu describe()

# 2. Configurar el estilo visual (para que se vea profesional)
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# ==========================================
# GRÁFICO 1: Gráfico de Barras (Comparación Directa)
# ==========================================
plt.figure()
ax = sns.barplot(x=sectores, y=promedios, palette=['#e74c3c', '#2ecc71'])

# Añadir títulos y etiquetas
plt.title('Promedio de Accidentes Laborales (Periodo 16 Años)\nConstrucción vs Suministro de Energía', fontsize=14, fontweight='bold')
plt.ylabel('Promedio de Accidentes', fontsize=12)
plt.xlabel('Sector Económico', fontsize=12)

# Añadir los números encima de las barras para mayor claridad
for i, v in enumerate(promedios):
    ax.text(i, v + 5, str(v), ha='center', fontweight='bold', fontsize=11)

plt.show() # Mostrar el primer gráfico

# ==========================================
# GRÁFICO 2: Gráfico de Pastel (Proporción de Riesgo)
# ==========================================
plt.figure(figsize=(8, 8))
# Usamos explode para separar el sector de energía y destacarlo
plt.pie(promedios, labels=sectores, autopct='%1.1f%%', startangle=140, 
        colors=['#e74c3c', '#2ecc71'], explode=(0, 0.1), shadow=True)

plt.title('Proporción de Riesgo entre Sectores\n(Basado en Accidentes Promedio)', fontsize=14, fontweight='bold')

plt.show() # Mostrar el segundo gráfico
```

```python
# Extraemos todas las categorías únicas de tu dataset limpio
categorias_unicas = df_long['Categoria_Sector_Sexo'].unique()

# Las guardamos en un archivo de texto en tu carpeta processed
ruta_txt = '../data/processed/lista_categorias.txt'

with open(ruta_txt, 'w', encoding='utf-8') as archivo:
    for categoria in categorias_unicas:
        archivo.write(f"{categoria}\n")

print(f"¡Listo! Se han exportado {len(categorias_unicas)} categorías al archivo: {ruta_txt}")
```

```python
# 1. Definir las dos categorías exactas que queremos comparar
categoria_hombres = 'Total con baja - Total - Hombres - Total - Accidentes'
categoria_mujeres = 'Total con baja - Total - Mujeres - Total - Accidentes'

# 2. Filtrar el dataframe largo para quedarnos solo con estas dos líneas
df_genero = df_long[df_long['Categoria_Sector_Sexo'].isin([categoria_hombres, categoria_mujeres])].copy()

# 3. Limpiar los nombres para que la leyenda del gráfico no sea gigante
df_genero['Categoria_Sector_Sexo'] = df_genero['Categoria_Sector_Sexo'].replace({
    categoria_hombres: 'Hombres',
    categoria_mujeres: 'Mujeres'
})

# 4. Configurar el gráfico de líneas
sns.set_theme(style="darkgrid") # Un fondo distinto para darle variedad a tu PDF
plt.figure(figsize=(12, 6))

# Dibujar las líneas (x = Fecha, y = Accidentes, color = Género)
sns.lineplot(
    data=df_genero, 
    x='Fecha', 
    y='Total_Accidentes', 
    hue='Categoria_Sector_Sexo', 
    palette=['#2980b9', '#9b59b6'], # Azul para hombres, Morado para mujeres
    linewidth=2.5
)

# 5. Títulos y etiquetas
plt.title('Evolución Histórica de Accidentes Laborales por Sexo', fontsize=15, fontweight='bold')
plt.ylabel('Total de Accidentes', fontsize=12)
plt.xlabel('Año', fontsize=12)
plt.legend(title='Sexo')
plt.tight_layout()

# --- NUEVO: Exportar a PNG ---
# OJO: Siempre debe ir antes de plt.show()
ruta_png = '../reports/evolucion_accidentes_sexo.png'
plt.savefig(ruta_png, dpi=300, bbox_inches='tight')
print(f"Gráfico exportado correctamente a: {ruta_png}")

# Mostrar el gráfico en el notebook
plt.show()
```
